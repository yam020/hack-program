Execute the code cell below to see an example of how it works.

```python
def random_words_api(nwords=10):
    "no docstring"
    URL = "https://random-word-api.herokuapp.com/word"
    response = requests.get(url=URL, params={"number": nwords})
    return response.json()

# demonstration
random_words_api(5)
```

### Challenge 2: 
Use the `random_words_api` function to get 50 random words and store the result as a variable. Write a function that takes the list of words as input and returns a dictionary with the longest word as the key and the length of the longest word as a value. If there is a tie in the length of words then have it return additional words as keys with their lengths as values.

```python
# get a list of words
words = random_words_api(50)
# function
def longestword(list):
    "store the longest words as the key with their lengths as values based on the input list"
    length = len(list[0])
    store = {}
    # get the longest length
    for i in range(len(list)):
        if len(list[i]) > length:
            length = len(list[i])
    # if the words mathes the same length, record it 
    for i in range(len(list)):
        if len(list[i]) == length:
            # ignore the repeated word
            if list[i] not in store:
                store[list[i]] = len(list[i])
    return store
```

```python
longestword(words)
```

### Challenge 3: 
Write a function to take the list of words as input and trim all words to be at most 5 characters in length, and return as a list.

```python
def trim_five(list):
    "return a list of trimmed words"
    newlist = []
    for word in list:
        # if words are longer than 5 characters, trim it down
        if len(word) >= 5:
            newlist.append(word[:5])
        # else do nothing
        else:
            newlist.append(word)
    return newlist
```

```python
trim_five(words)
```

### Challenge 4: 
Write a function to take a list of words as input and to count the occurrence of all letters in every word and return as a dictionary mapping letters to integers., e.g., {'a': 10, 'b': 3, 'c': 5, ...}.  

```python
def count(list):
    "count every unique letters"
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    counter = {}
    for char in alphabet:
        counter[char] = 0
    # nested loop to get every letter
    for word in list:
        for letter in word:
            # map each unique letter
            counter[letter.lower()] += 1
    return counter
```

```python
count(words)
```

### Challenge 6: 
Using numpy create a new variable called `arr` with 1000 random samples from a normal distribution. Use the numpy `.histogram` function to bin these values into 20 bins, and then plot the histogram using a barplot from toyplot. Color the bars of the histogram orange.

```python
# Create a new variable with 1000 random samples from a normal distribution 
# The mean is 0, and the standard deviation is 1
arr = np.random.normal(loc=0, scale=1, size=1000)
# bin the above arr into 20 bins
hist, edges = np.histogram(arr, bins=20, density=True)
# chose the color orange
color = toyplot.color.css("orange")
# draw the plot 
canvas = toyplot.Canvas(width=400, height=200)
axes = canvas.cartesian(label="Normal Distributed Random Values")
bars = axes.bars((hist, edges), color = color)
```

```python
# simple version without the recoginizing types of the words
def simmadlib():
    words = random_words_api(4)
    print( f"""
    -- Do you know the meaning of {words[0]}?
    -- No.
    -- How about {words[1]}? 
    -- I dunno.
    -- Ohhh, how about {words[2]}?
    -- I have no idea ...
    -- Okay... What about this one "{words[3]}"?
    -- Go away!
    """ )
simmadlib()
```

```python
# install the nltk natural language library nlkt and import the library: nltk
import nltk
from nltk.tokenize import word_tokenize
from numpy import random
# random picking words
def typewd(dfwords,type):
    grouped = dfwords.groupby("Type") 
    df_type = grouped.get_group(type)
    word = df_type.sample().iat[0,0]
    return word
# use the package to understand the type of words 
def madlib_upgrade():
    df_new = None
    while 1:
        words = ' '.join(random_words_api(50))
        words = word_tokenize(words)
        word_list = nltk.pos_tag(words,tagset="universal")
        dfwords = pd.DataFrame(word_list, columns = ['Word', 'Type'])
        s = pd.Series(['NOUN', 'ADJ', 'ADV', 'VERB'])
        if s.isin(dfwords['Type']).all():
            df_new = dfwords
            break
    Noun = typewd(df_new,"NOUN")
    Adj = typewd(df_new,"ADJ")
    Verb = typewd(df_new,"VERB")
    Adv = typewd(df_new,"ADV")
    print(f"""
    Today is a {Adj} day. I {Adv} {Verb} a {Noun}. 
    """)
```

```python
madlib_upgrade()
```
